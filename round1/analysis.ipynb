```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim

# Prepare data for LSTM
def prepare_data(data, lookback=60):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data.reshape(-1, 1))
    
    X, y = [], []
    for i in range(len(scaled_data) - lookback):
        X.append(scaled_data[i:(i + lookback), 0])
        y.append(scaled_data[i + lookback, 0])
    
    return np.array(X), np.array(y), scaler

# Get KELP price data
kelp_prices = KELP['mid_price'].values

# Prepare sequences
X, y, scaler = prepare_data(kelp_prices)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Convert to PyTorch tensors
X_train = torch.FloatTensor(X_train).unsqueeze(-1)
X_test = torch.FloatTensor(X_test).unsqueeze(-1)
y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)
```

```python

# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=2, output_size=1):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        
        out, _ = self.lstm(x, (h0, c0))
        out = self.dropout(out[:, -1, :])
        out = self.fc(out)
        return out

# Initialize model
model = LSTMModel()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters())

# Train model
num_epochs = 25
batch_size = 32

for epoch in range(num_epochs):
    model.train()
    for i in range(0, len(X_train), batch_size):
        batch_X = X_train[i:i+batch_size]
        batch_y = y_train[i:i+batch_size]
        
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs.squeeze(), batch_y)
        loss.backward()
        optimizer.step()

    # Print progress bar
    progress = (epoch + 1) / num_epochs
    bar_length = 30
    filled_length = int(bar_length * progress)
    bar = '=' * filled_length + '-' * (bar_length - filled_length)
    print(f'\rTraining Progress: [{bar}] {progress:.1%}', end='')
    
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python

# Make predictions
model.eval()
with torch.no_grad():
    train_predict = model(X_train).numpy()
    test_predict = model(X_test).numpy()

# Inverse transform predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train_actual = scaler.inverse_transform(y_train.numpy().reshape(-1, 1))
y_test_actual = scaler.inverse_transform(y_test.numpy().reshape(-1, 1))

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(KELP['time'][-len(y_test_actual):], y_test_actual, label='Actual')
plt.plot(KELP['time'][-len(test_predict):], test_predict, label='Predicted')
plt.title('KELP Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Calculate RMSE
rmse = np.sqrt(np.mean((test_predict - y_test_actual) ** 2))
print(f'Test RMSE: {rmse:.2f}')

```

```python
# Calculate MACD
def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    # Calculate short and long term EMAs
    short_ema = data['mid_price'].ewm(span=short_window, adjust=False).mean()
    long_ema = data['mid_price'].ewm(span=long_window, adjust=False).mean()
    
    # Calculate MACD line
    macd_line = short_ema - long_ema
    
    # Calculate signal line
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()
    
    # Calculate MACD histogram
    macd_histogram = macd_line - signal_line
    
    return macd_line, signal_line, macd_histogram

# Calculate RSI
def calculate_rsi(data, periods=14):
    delta = data['mid_price'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=periods).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=periods).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# Calculate support and resistance levels
def find_support_resistance(data, window=20):
    rolling_min = data['mid_price'].rolling(window=window).min()
    rolling_max = data['mid_price'].rolling(window=window).max()
    return rolling_min, rolling_max

# Calculate indicators
macd_line, signal_line, macd_histogram = calculate_macd(KELP)
rsi = calculate_rsi(KELP)
support, resistance = find_support_resistance(KELP)

# Plot MACD
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(KELP['time'], KELP['mid_price'], label='Price')
plt.plot(KELP['time'], support, 'g--', label='Support', alpha=0.5)
plt.plot(KELP['time'], resistance, 'r--', label='Resistance', alpha=0.5)
plt.title('KELP Price with Support/Resistance')
plt.legend()
plt.grid(True)

# Plot MACD
plt.subplot(3, 1, 2)
plt.plot(KELP['time'], macd_line, label='MACD')
plt.plot(KELP['time'], signal_line, label='Signal')
plt.bar(KELP['time'], macd_histogram, label='Histogram', alpha=0.3)
plt.title('MACD')
plt.legend()
plt.grid(True)

# Plot RSI
plt.subplot(3, 1, 3)
plt.plot(KELP['time'], rsi, label='RSI')
plt.axhline(y=70, color='r', linestyle='--', label='Overbought')
plt.axhline(y=30, color='g', linestyle='--', label='Oversold')
plt.title('RSI')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Print some statistics
print("\nRSI Statistics:")
print(f"Mean RSI: {rsi.mean():.2f}")
print(f"Max RSI: {rsi.max():.2f}")
print(f"Min RSI: {rsi.min():.2f}")
print(f"Number of overbought periods: {len(rsi[rsi > 70])}")
print(f"Number of oversold periods: {len(rsi[rsi < 30])}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Buy/Sell Signal and Profit Calculation Logic ---

def generate_macd_signals(data, macd_line, signal_line):
    buy_signals, sell_signals = [], []
    position = False
    buy_price = 0
    profits = []

    for i in range(1, len(data)):
        if macd_line.iloc[i] > signal_line.iloc[i] and macd_line.iloc[i - 1] <= signal_line.iloc[i - 1]:
            if not position:
                buy_signals.append(data['mid_price'].iloc[i])
                sell_signals.append(None)
                buy_price = data['mid_price'].iloc[i]
                position = True
            else:
                buy_signals.append(None)
                sell_signals.append(None)
        elif macd_line.iloc[i] < signal_line.iloc[i] and macd_line.iloc[i - 1] >= signal_line.iloc[i - 1]:
            if position:
                sell_signals.append(data['mid_price'].iloc[i])
                buy_signals.append(None)
                profits.append(data['mid_price'].iloc[i] - buy_price)
                position = False
            else:
                buy_signals.append(None)
                sell_signals.append(None)
        else:
            buy_signals.append(None)
            sell_signals.append(None)

    # Add initial None for alignment
    buy_signals.insert(0, None)
    sell_signals.insert(0, None)

    return buy_signals, sell_signals, profits


def generate_rsi_signals(data, rsi, overbought=70, oversold=30):
    buy_signals, sell_signals = [], []
    position = False
    buy_price = 0
    profits = []

    for i in range(1, len(rsi)):
        rsi_now = rsi.iloc[i]
        rsi_prev = rsi.iloc[i - 1]

        if rsi_now < oversold and rsi_prev >= oversold:
            if not position:
                buy_signals.append(data['mid_price'].iloc[i])
                sell_signals.append(None)
                buy_price = data['mid_price'].iloc[i]
                position = True
            else:
                buy_signals.append(None)
                sell_signals.append(None)

        elif rsi_now > overbought and rsi_prev <= overbought:
            if position:
                sell_signals.append(data['mid_price'].iloc[i])
                buy_signals.append(None)
                profits.append(data['mid_price'].iloc[i] - buy_price)
                position = False
            else:
                buy_signals.append(None)
                sell_signals.append(None)
        else:
            buy_signals.append(None)
            sell_signals.append(None)

    # Pad the beginning with None
    buy_signals.insert(0, None)
    sell_signals.insert(0, None)

    return buy_signals, sell_signals, profits

def generate_support_resistance_signals(data, support, resistance):
    buy_signals, sell_signals = [], []
    position = False
    buy_price = 0
    profits = []

    for i in range(len(data)):
        price = data['mid_price'].iloc[i]
        support_level = support.iloc[i]
        resistance_level = resistance.iloc[i]

        if price <= support_level:
            if not position:
                buy_signals.append(price)
                sell_signals.append(None)
                buy_price = price
                position = True
            else:
                buy_signals.append(None)
                sell_signals.append(None)
        elif price >= resistance_level:
            if position:
                sell_signals.append(price)
                buy_signals.append(None)
                profits.append(price - buy_price)
                position = False
            else:
                buy_signals.append(None)
                sell_signals.append(None)
        else:
            buy_signals.append(None)
            sell_signals.append(None)

    return buy_signals, sell_signals, profits



# --- Apply Strategies ---
macd_buys, macd_sells, macd_profits = generate_macd_signals(KELP, macd_line, signal_line)
rsi_buys, rsi_sells, rsi_profits = generate_rsi_signals(KELP, rsi)
sr_buys, sr_sells, sr_profits = generate_support_resistance_signals(KELP, support, resistance)

# --- Print Profits ---
print("\n--- Strategy Profits ---")
print(f"MACD Total Profit: {sum(macd_profits):.2f} over {len(macd_profits)} trades")
print(f"RSI Total Profit: {sum(rsi_profits):.2f} over {len(rsi_profits)} trades")
print(f"Support/Resistance Total Profit: {sum(sr_profits):.2f} over {len(sr_profits)} trades")

# --- Optional: Plot Buy/Sell Points ---
def plot_signals(data, buy_signals, sell_signals, title, xlim=None):
    plt.figure(figsize=(10, 4))
    plt.plot(data['time'], data['mid_price'], label='Price')
    plt.scatter(data['time'], buy_signals, label='Buy', marker='^', color='g')
    plt.scatter(data['time'], sell_signals, label='Sell', marker='v', color='r')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    if xlim != None:
        plt.xlim(xlim[0], xlim[1])
    plt.show()

plot_signals(KELP, macd_buys, macd_sells, 'MACD Buy/Sell Signals')
plot_signals(KELP, rsi_buys, rsi_sells, 'RSI Buy/Sell Signals')
plot_signals(KELP, sr_buys, sr_sells, 'Support/Resistance Buy/Sell Signals')
```

```python
def generate_trailing_stop_signals(data, lookback=30, stop_loss_pct=0.02, max_hold_period=1000):
    buy_signals = [None] * len(data)
    sell_signals = [None] * len(data)
    profits = []
    active_positions = []

    for i in range(lookback, len(data)):
        price = data['mid_price'].iloc[i]
        lookback_prices = data['mid_price'].iloc[i-lookback:i]

        # Open new position if price is at the 20-period low
        if price == min(lookback_prices):
            position = {
                "buy_price": price,
                "stop_loss": price * (1 - stop_loss_pct),
                "entry_index": i,
                "entry_price": price
            }
            active_positions.append(position)
            buy_signals[i] = price

        # Check all active positions
        updated_positions = []
        for position in active_positions:
            entry_index = position["entry_index"]
            buy_price = position["buy_price"]
            stop_loss = position["stop_loss"]

            # Update stop loss if price has increased
            if price > buy_price:
                new_stop = price * (1 - stop_loss_pct)
                position["stop_loss"] = max(stop_loss, new_stop)

            # Check if stop loss is hit
            if price <= position["stop_loss"]:
                sell_signals[i] = price
                profits.append(price - buy_price)
            # Check if position has expired
            elif i - entry_index >= max_hold_period:
                sell_signals[i] = price
                profits.append(price - buy_price)
            else:
                updated_positions.append(position)

        active_positions = updated_positions

    # Final step: close any remaining positions at the last price
    final_price = data['mid_price'].iloc[-1]
    print("Active Positions: ", len(active_positions))
    for position in active_positions:
        entry_index = position["entry_index"]
        buy_price = position["buy_price"]
        sell_signals[-1] = final_price
        profits.append(final_price - buy_price)

    return buy_signals, sell_signals, profits

# Apply trailing stop strategy
ts_buys, ts_sells, ts_profits = generate_trailing_stop_signals(KELP)

# Print profits
print(f"\nTrailing Stop Strategy Total Profit: {sum(ts_profits):.2f} over {len(ts_profits)} trades")

# Plot signals
plot_signals(KELP, ts_buys, ts_sells, 'Trailing Stop Buy/Sell Signals')
```
